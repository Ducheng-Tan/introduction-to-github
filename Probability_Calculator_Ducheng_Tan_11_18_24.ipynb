Formulas:

Geometric Distribution:


$$\forall x \in \mathbb{N}^+$$

$$P(X = x) = (1-p)^{x-1}p $$

$$P(X \leq x) =  1- (1-p)^x $$

$$P(X \geq x) = (1-p)^{k-1}$$

$$ E[X] = \frac{1}{p} $$

$$ Var[x] =  \frac{1-p}{p^2} $$


Binomial Distribution:

$$\forall x \in \mathbb{N}^+$$

$$P(X=x) = {{n}\choose{x}}p^x(1-p)^{n-x} $$

$$ P(X\leq x) = \sum_{x=1}^{n} {{n}\choose{x}}p^x(1-p)^{n-x}$$

$$P(X\geq x) = 1- \left[ \sum_{x=1}^{n} {{n}\choose{x}}p^x(1-p)^{n-x}\right]$$

$$ E[X] = {np} $$

$$ Var[x] =  np(1-p) $$



*NOTE: The CDF from the code only calculates $P(X≤x)$ , to compute $P(X≥x)$, Compute $1- P(X≤x)$.
$$  P(X≥x) = 1- P(X≤x)$$

***The CODE/The Probability Calculator ***

```python
import math
# FIrst defining all the equations and assigning formulas to them
def geometric_pdf(x, p):
    """Calculates PDF for Geometric distribution."""
    return (1 - p) ** (x - 1) * p

def geometric_cdf(x, p):
    """Calculates CDF for Geometric distribution."""
    return sum(geometric_pdf(k, p) for k in range(1, x + 1))

def geometric_expected_value(p):
    """Calculates the expected value for Geometric distribution."""
    return 1 / p

def geometric_variance(p):
    """Calculates the variance for Geometric distribution."""
    return (1 - p) / (p**2)

def binomial_pdf(r, n, p):
    """Calculates PDF for Binomial distribution."""
    return math.comb(n, r) * (p ** r) * ((1 - p) ** (n - r))

def binomial_cdf(r, n, p):
    """Calculates CDF for Binomial distribution."""
    return sum(binomial_pdf(k, n, p) for k in range(r + 1))

def binomial_expected_value(n, p):
    """Calculates the expected value for Binomial distribution."""
    return n * p

def binomial_variance(n, p):
    """Calculates the variance for Binomial distribution."""
    return n * p * (1 - p)


# defining a function that allows user interaction
def main():
    # Select distribution type
    print("Choose Distribution Type:\n")
    print("1. Geometric")
    print("2. Binomial")
    dist_choice = input("Enter Geometric for Geometric or Binomial for Binomial: \n")

    # Geometric Distribution
    if dist_choice == "Geometric":
        p = float(input("Enter probability of success (0 <= p <= 1): \n "))
        if not (0 <= p <= 1):
            print("Probability must be between 0 and 1.")
            return

        print("Choose Calculation Type:\n")
        print("1. PDF (Probability Density Function)")
        print("2. CDF (Cumulative Distribution Function)")
        calc_type = input("Enter 1 for PDF or 2 for CDF: \n")

        if calc_type == "1":  # PDF
            x = int(input("Enter the trial number (x) where success occurs:\n"))
            result = geometric_pdf(x, p)
            print(f"Geometric PDF at x={x}: {result}")
        elif calc_type == "2":  # CDF
            x = int(input("Enter the number of trials (x): \n"))
            result = geometric_cdf(x, p)
            print(f"Geometric CDF for x<={x}: {result}")

        # Calculate and print expected value and variance after user input
        expected_value = geometric_expected_value(p)
        variance = geometric_variance(p)
        print(f"Geometric Expected Value: {expected_value}")
        print(f"Geometric Variance: {variance}")

    # Binomial Distribution
    elif dist_choice == "Binomial":
        p = float(input("Enter probability of success (0 <= p <= 1):\n "))
        if not (0 <= p <= 1):
            print("Probability must be between 0 and 1.")
            return
        n = int(input("Enter the number of trials (n): \n"))

        print("Choose Calculation Type:\n")
        print("1. PDF (Probability Density Function)")
        print("2. CDF (Cumulative Distribution Function)")
        calc_type = input("Enter 1 for PDF or 2 for CDF: \n")

        if calc_type == "1":  # PDF
            r = int(input("Enter the number of successes (r):\n "))
            result = binomial_pdf(r, n, p)
            print(f"Binomial PDF for r={r}: {result}")
        elif calc_type == "2":  # CDF
            r = int(input("Enter the number of successes (r) for cumulative probability: \n"))
            result = binomial_cdf(r, n, p)
            print(f"Binomial CDF for r<={r}: {result}")

        # Calculate and print expected value and variance after user input
        expected_value = binomial_expected_value(n, p)
        variance = binomial_variance(n, p)
        print(f"Binomial Expected Value: {expected_value}")
        print(f"Binomial Variance: {variance}")

    else:
        print("Invalid distribution choice.")

main()
```
